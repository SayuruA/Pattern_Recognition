```python
import numpy as np
import matplotlib.pyplot as plt
import numpy as np
from sklearn.datasets import make_blobs
# Generate synthetic data
np.random.seed(0)
centers = [[-5, 0], [5, 1.5]]
X, y = make_blobs(n_samples=2000, centers=centers, random_state=5)
transformation = [[0.5, 0.5], [-0.5, 1.5]]
X = np.dot(X, transformation)
```

```python
def sigmoid(z):
  return 1/(1+np.exp(-z))
```

```python
def gradient(x_batch, y_batch, w):
  """
This takes a batch of data points and returns the gradient of the logistic regression model

  """

  sum_grads = np.zeros(w.shape)

  for j in range(len(x_batch)):
    x = x_batch[j] # 1*2
    x = np.append(x, 1) # 1*3
    temp1 = sigmoid( np.dot(w, x)) - y_batch[j] # scalar
    sum_grads += temp1 * x

  sum_grads /= len(x_batch)

  return sum_grads
```

```python
def accuracy(x_batch, y_batch, w):
  """
This takes a batch of data points and returns the accuracy of the logistic regression model
  """
  y_pred = np.matmul(x_batch, np.transpose(w[1:])) + w[0]
  y_pred = sigmoid(y_pred)
  y_pred = np.where(y_pred > 0.5, 1, 0)
  return np.mean(y_pred == y_batch)
```

```python
def update_weights(X, y, w, a, b_size):

  # go through each batch
  for i in range(0, X.shape[0], b_size):
    x_batch = X[i:i+b_size] # b_size * 2
    y_batch = y[i:i+b_size] # 1 * b_size
    grad = gradient(x_batch, y_batch, w)
    # update weights
    w -= a * grad
    #if i == 0:print(grad)
  #print(grad)
  return w, grad
```

```python
def shuffle_data(X, y):
  Xy = np.concatenate((X, y.reshape(-1,1)), axis=1)
  np.random.shuffle(Xy)
  X = Xy[:, :-1]
  y = Xy[:, -1]
  return X, y
```

```python
def NLL(X, y, w):
  sigma = sigmoid(np.matmul(X, np.transpose(w[1:])) + w[0])
  log_sigma = np.log(sigma)
  log_1_sigma = np.log(1 - sigma)
  NLL_error = -np.mean(y * log_sigma + (1 - y) * log_1_sigma)
  return NLL_error
```

```python
def logiReg_batch(X, y, a = 0.01, b_size = 100, num_epochs = 200):

  # batch size
  #b_size = 100

  # alpha
  # a = 0.001

  # number of epochs
  # num_epochs = 20

  # initialize weights vector
  # random
  # w = np.random.randn(X.shape[1])*2
  # w = np.append(w, 1)

  # arbitrary
  # w = np.array([ 2.72195146,  1.70157852, -0.49984544],dtype = np.float64)

  # zeros
  w = np.zeros(X.shape[1]+1) # 1*3
  # store the loss after each epoch
  loss = np.zeros(num_epochs)

  for epoch in range(num_epochs):

    if epoch > 200:
      a = 0.01
    # update w through all batches
    w, grad = update_weights(X, y, w, a, b_size)


    # shuffle the data after every epoch
    X, y = shuffle_data(X, y)
    L = NLL(X, y, w)
    loss[epoch] = L
    # print("Loss at epoch", epoch , "=", NLL(X, y, w))
  print("final gradient = ", grad)
  return w, loss
```

```python
Accuracy = 0
while Accuracy < 0.9:
  w, loss = logiReg_batch(X, y, num_epochs = 20)
  Accuracy = accuracy(X, y, w)
  print(Accuracy)


# plot the loss
fig, ax = plt.subplots()
ax.plot(loss)
ax.set_xlabel('Epoch')
ax.set_ylabel('Loss')
ax.set_title('Loss vs Epoch')
ax.grid(True)
plt.show()
```

```python
import numpy as np

def sigmoid(z):
    return 1 / (1 + np.exp(-z))

def newton_method_logistic_regression(X, y, max_iter=10):
    # Add bias term (intercept) to X
    X = np.c_[np.ones(X.shape[0]), X]

    # Initialize weights
    w = np.random.randn(X.shape[1])*2

    for iteration in range(max_iter):
        # Predicted probabilities
        h = sigmoid(np.dot(X, w))

        # Gradient (X^T * (h - y))
        gradient = np.dot(X.T, (h - y))

        # Diagonal matrix R (h(x) * (1 - h(x)))
        R = np.diag(h * (1 - h))

        # Hessian (X^T * R * X)
        H = np.dot(np.dot(X.T, R), X)

        # Update weights using Newton's method
        w -= np.linalg.inv(H).dot(gradient)

    return w
```

```python
w = newton_method_logistic_regression(X, y)
Accuracy = accuracy(X, y, w)
print(Accuracy)
```
